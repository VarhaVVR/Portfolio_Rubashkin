# 2. Создание вспомогательных функций

```python
#Определяю необходимые функции
#Отображение графика для признака, графика логарифмированного признака и boxplot для определения выбросов
def plotting (col):
    fig, ax = plt.subplots(3,1, figsize=(10,9))
    sns.distplot(col,ax = ax[0],kde = False, rug=False)
    sns.distplot(col.apply(lambda x: m.log(x+1) if x >= 0 else m.log(-x)), ax = ax[1] ,kde = False, rug = False)
    sns.boxplot(x = col,ax = ax[2], showmeans = True)
    ax[0].set_title('Гистограмма признака {}'.format(col.name), fontsize = 14)
    ax[1].set_title('Гистограмма логарифмированного признака {}'.format(col.name), fontsize = 14)
    ax[2].set_title('Boxplot признака {}'.format(col.name), fontsize = 14)
    fig.tight_layout(pad = 3)
    return

# Определение количества выбросов для признака
def outlier_check(column):
    Q1 = np.percentile(column, 25)
    Q3 = np.percentile(column, 75)
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    print('Нижняя граница: {}'.format(round(lower_range,2)),
          'Верняя граница: {}'.format(round(upper_range,2)),
          'Количество выбросов слева: {}'.format(len(data[column<lower_range])),
          'Количество выбросов справа: {}'.format(len(data[column>upper_range])), 
          sep='\n')

# Отображение confusion matrix 
def show_confusion_matrix(y_true, y_pred):
    color_text = plt.get_cmap('BrBG')(1.0)
    class_names = ['Дефолт', 'Не дефолт']
    cm = confusion_matrix(y_true, y_pred)
    cm[0,0], cm[1,1] = cm[1,1], cm[0,0]
    df = pd.DataFrame(cm, index=class_names, columns=class_names)
    
    fig, ax = plt.subplots(figsize=(10, 10))
    ax.set(xticks=np.arange(cm.shape[1]), yticks=np.arange(cm.shape[0]), title="Матрица ошибок")
    ax.title.set_fontsize(20)
    sns.heatmap(df, square=True, annot=True, fmt="d", linewidths=1, cmap="BrBG",annot_kws={"size": 18})
    plt.setp(ax.get_yticklabels(), rotation=0, ha="right", rotation_mode="anchor", fontsize=20)
    plt.setp(ax.get_xticklabels(), rotation=0, ha="center", rotation_mode="anchor", fontsize=20)
    ax.set_ylabel('Предсказанные значения', fontsize=20, color = color_text)
    ax.set_xlabel('Реальные значения', fontsize=20, color = color_text)
    b, t = plt.ylim()
    plt.ylim(b+0.1, t+0.1)
    fig.tight_layout()
    plt.show()

# Функция отображения целевой метрики
def show_roc_curve(y_true, y_pred_prob):
    fpr, tpr, _ = roc_curve(y_true, y_pred_prob)
    plt.figure()
    plt.plot([0, 1], label='Случайный классификатор', linestyle='--')
    plt.plot(fpr, tpr, label = 'Логистическая регрессия')
    plt.title('Логистическая регрессия ROC AUC = %0.5f' % roc_auc_score(y_true, y_pred_prob))
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.legend(loc = 'lower right')
    plt.show()
    
# Построение кривой precision-recall    
def recall_precision_plot(y_test, y_pred):
    precisions, recalls, _ = precision_recall_curve(y_test, y_pred)
    ap = average_precision_score(y_test, y_pred)        
   
    plt.figure()
       
    plt.step(recalls, precisions, color='b', alpha=0.2, where='post')
    plt.fill_between(recalls, precisions, step='post', alpha=0.2, color='lime')
    plt.xlabel('Recall');
    plt.ylabel('Precision');
    plt.title('Recall-precision curve, площадь под кривой = %0.10f' % ap)
    plt.grid(True)
    plt.show()
```

# 5. Подготовка модели

```python
# Данные для модели
train_data = data.query('train == 1').drop(['train','client_id','app_date'], axis=1)
test_data = data.query('train == 0').drop(['train','default','app_date'], axis=1)
```

```python
X = train_data.drop(['default'],axis = 1)
y = train_data['default']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y , test_size=0.20, random_state=random_seed)
```

# 6. Модели
## 6.1 LogisticRegresion without hyperpameters tweaking

```python
model = LogisticRegression(max_iter = 2000, random_state = random_seed)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
probs = model.predict_proba(X_test)
probs = probs[:,1]


fpr, tpr, threshold = roc_curve(y_test, probs)
roc_auc = roc_auc_score(y_test, probs)

show_roc_curve(y_test, probs)
show_confusion_matrix(y_test, y_pred)
recall_precision_plot(y_test, y_pred)
```

Значение ключевой метрики достаточно высокое, однако модель обладает низкой предсказательной способностью, так как плохо определяет дефолтных клиентов. Попробую построить модель со сбалансированной выборкой.


## 6.2 LogisticRegresion with balanced class weight


```python
model2 = LogisticRegression(class_weight='balanced', max_iter = 2000, random_state = random_seed)
model2.fit(X_train, y_train)
y_pred2 = model2.predict(X_test)
probs2 = model2.predict_proba(X_test)
probs2 = probs2[:,1]


fpr, tpr, threshold = roc_curve(y_test, probs2)
roc_auc = roc_auc_score(y_test, probs2)

show_roc_curve(y_test, probs2)
show_confusion_matrix(y_test, y_pred2)
recall_precision_plot(y_test, y_pred2)
```

Хоть и значения ключевой метрики не изменились, модель стала гораздо точнее предсказывать дефолтных клиентов, уменьшилось количество ошибок первого рода, но увеличилось количество ошибок второго. также немного подняли значения по recall-precision. Самое время поиграть гиперпараметрами.

## 6.3 LogisticRegresion with GridSearch

```python
# Определяем параметры для подбора
C = np.logspace(-2, 2, 10)
iter_ = 100
epsilon_stop = 1e-4
 
hyperparameters = [
    {'penalty': ['l1'], 
     'C': C,
     'solver': ['liblinear', 'saga'], 
     'multi_class': ['auto','ovr'], 
     'max_iter':[iter_],
     'tol':[epsilon_stop]},
    {'penalty': ['l2'], 
     'C': C,
     'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'], 
     'multi_class': ['auto','ovr'], 
     'max_iter':[iter_],
     'tol':[epsilon_stop]},
    {'penalty': ['none'], 
     'C': C,
     'solver': ['newton-cg', 'lbfgs', 'sag', 'saga'], 
     'multi_class': ['auto','ovr'], 
     'max_iter':[iter_],
     'tol':[epsilon_stop]},
    {'penalty': ['elasticnet'], 
     'C': C,
     'solver': ['saga'], 
     'multi_class': ['auto','ovr'], 
     'max_iter':[iter_],
     'tol':[epsilon_stop]},
]

gridsearch = GridSearchCV(model2, hyperparameters, scoring='roc_auc', n_jobs=-1, cv=5)
gridsearch.fit(X_train, y_train)
model = gridsearch.best_estimator_

best_parameters = model.get_params()
for param_name in sorted(best_parameters.keys()):
        print('\t%s: %r' % (param_name, best_parameters[param_name]))
```

```python
model3 = LogisticRegression(penalty = 'l2', 
                            dual = False, 
                            tol = 0.0001,
                            C = 0.01, 
                            fit_intercept = True,
                            intercept_scaling = 1,
                            class_weight = 'balanced',
                            random_state= 42,
                            solver= 'newton-cg',
                            max_iter= 100,
                            multi_class='auto',
                            verbose=0,
                            warm_start=False)
                                                     


model3.fit(X_train, y_train)
y_pred3 = model3.predict(X_test)
probs3 = model3.predict_proba(X_test)
probs3 = probs3[:,1]


fpr, tpr, threshold = roc_curve(y_test, probs3)
roc_auc = roc_auc_score(y_test, probs3)

show_roc_curve(y_test, probs3)
show_confusion_matrix(y_test, y_pred3)
recall_precision_plot(y_test, y_pred3)
```

## 6.5 Метод undersampling для балансировки определяемого признака.

```python
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(random_state=42)
X_us, y_us = rus.fit_resample(X, y)
X_train, X_test, y_train, y_test = train_test_split(X_us, y_us, test_size=0.20, shuffle = True, random_state=random_seed)
```

```python
Cs_ = np.logspace(-4, 4, 100)
model5 = LogisticRegressionCV(
  Cs = Cs_,
  cv = 5,
  scoring = 'roc_auc',
  solver = 'saga',
  tol = 0.0001,
  max_iter = 1000,   
  n_jobs = -1,
  random_state = 42,
)

model5.fit(X_train, y_train)

y_pred5 = model5.predict(X_test)
probs5 = model5.predict_proba(X_test)
probs5 = probs5[:,1]


fpr, tpr, threshold = roc_curve(y_test, probs5)
roc_auc = roc_auc_score(y_test, probs5)
show_roc_curve(y_test, probs5)
show_confusion_matrix(y_test, y_pred5)
recall_precision_plot(y_test, y_pred5)
```

Модель потеряла в конечной метрике roc_auc, но зато  в 3 раза возрос средний показатель recall-precision, что может говорить о том, что модель стала точнее определять нужный признак и roc-auc не самая лучшая характеристика для определения работоспособности модели бинарного классификатора в данном случае.
