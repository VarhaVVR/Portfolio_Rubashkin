## Задача предсказания цены автомобиля на основании спарсенных объявлений с сайта auto.ru.
## Задача включает в себя парсинг данных, их обработку, создание дополнительных фич, и предсказание на нескольких моделях, для получения максимальной точности по MAPE.

```python
import pandas as pd
import numpy as np
import statistics as s
from sklearn.model_selection import train_test_split
from catboost import CatBoostRegressor
from catboost import cv
from catboost import Pool
import math as m
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import GridSearchCV
import seaborn as sns
import matplotlib.pyplot as plt
import joblib
from sklearn.preprocessing import StandardScaler
```

## Обработка спарсенных данных в другом файле. Этот notebook предназначен только для работ с алгоритмами.

```python
df = pd.read_csv(r'E:\5 project\car_df_v2.csv',index_col=[0])
```

```python
df.head().T
```

```python
df.isna().sum()
```

```python
df['bodyType'].fillna(df['bodyType'].value_counts().idxmax(), inplace = True)
df['fuelType'].fillna(df['fuelType'].value_counts().idxmax(), inplace = True)
df['ПТС'].fillna(df['ПТС'].value_counts().idxmax(), inplace = True)
df['Руль'].fillna(df['Руль'].value_counts().idxmax(), inplace = True)
```

```python
# Вспомогательные функции для обработки датафреймов
# Визуализация графика численной переменной до и после логарифмизации
def plotting (col):
    fig, ax = plt.subplots(2,1, figsize=(10,9))
    sns.distplot(col,ax = ax[0],kde = False, rug=False)
    sns.distplot(col.apply(lambda x: m.log(x+1) if x >= 0 else m.log(-x)), ax = ax[1] ,kde = False, rug = False)
    ax[0].set_title('Гистограмма признака {}'.format(col.name), fontsize = 14)
    ax[1].set_title('Гистограмма логарифмированного признака {}'.format(col.name), fontsize = 14)
    fig.tight_layout(pad = 3)
    return

# Cоздание полиномиальных признаков 3 степени. Признаки вида 1, a, b, c, ab, ac, abc, a**2b...
def poly_3(df,columns):
    pf = PolynomialFeatures(3)
    poly_features = pf.fit_transform(df[columns])
    poly = pd.DataFrame(poly_features)
    poly_dict={}
    for i in range(len(poly.columns)):
        poly_dict[i] = f'poly_{i}'
    poly.rename(columns = poly_dict,inplace = True)
    df.reset_index(drop = True,inplace = True)
    result = pd.concat([df, poly], axis=1)
#     Изначально была идея удалить сильно скореллированные признаки, но точность от этого упала.
#     for i in [0,2,8,11,3,4,12,13,14,5,15,6,17,18,19,16]:
#         result.drop([f'poly_{i}'],axis=1,inplace=True)
    return result
# Создание функции ошибки
def mape(y_true, y_pred):
    return np.mean(np.abs((y_pred-y_true)/y_true))
```

```python
plotting(df['price'])
```

```python
plotting(df['mileage'])
```

```python
# Логарифмирование численных переменных позволяет улучшить точность по MAPE с 32% до 18%
df['mileage'] = df['mileage'].apply(lambda x: m.log(x+1))
df['price'] = df['price'].apply(lambda x: m.log(x+1))
```

```python
df.groupby('brand')['enginePower'].mean().sort_values(ascending = False)[:10]
```

```python
# Добавляю столбец с флагом, обозначающим, что значение мощности больше среднего по группе.
df['big_HP'] = 0
df.loc[df['enginePower']>df.groupby('brand')['enginePower'].transform('mean'),['big_HP']]=1
```

```python
df.head().T
```

```python
df = df.drop_duplicates(subset = ['bodyType', 'brand', 'color', 'enginePower', 'fuelType', 'mileage',
       'modelDate', 'model_name', 'vehicleTransmission',
       'Владельцы', 'ПТС', 'Руль', 'train', 'warranty', 'big_HP'])
```

```python
# Признак популярного цвета, цвет популярен, если находится в топ-5 цветов.
df['popular_color'] = df['color']
popular_color = list(df['color'].value_counts()[:5].keys())
df['popular_color'] = df['popular_color'].apply(lambda x: 1 if x in popular_color else 0)
```

```python

result = poly_3(df,['enginePower','mileage'])
```

```python
correlation = result.corr()
plt.figure(figsize=(20, 20))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
```

```python
# Обозначаю категориальные переменные для CatBoost
cat_f = []
for item in result.columns:
    if type(result[item][0])== str:
        cat_f.append(item)
```

```python
# Кодирую категориальные переменные. В документации CatBoost указано, что в этом нет необходимости, 
# но без кодирования точность меньше на пару процентов.
for col in cat_f:
    result[col] = result[col].astype('category').cat.codes
```

```python
X = result[result['train']==1].drop(['price',],axis = 1)
X.drop(['train'],axis=1,inplace = True)
y = result[result['train']==1]['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False, random_state=42)
```

## Используемые модели

```python
## В предыдущей версbи были использованы Catboost, RandomForest, XGBoost, LightGBM Regressor, лучший результат показал CatBoost.
```

```python
# В прошлой версии был произведен поиск гиперпараметров с помощью GridSearchCV, но лучший результат показали дефолтные параметры модели.
model = CatBoostRegressor(iterations = 1000,
                            random_seed = 42,
                            eval_metric='MAPE',
                            )
model.fit(X_train, y_train,
                    cat_features=cat_f,
                    eval_set=(X_test, y_test),
                    verbose_eval=250,
                    use_best_model=True,
                    plot=True
                    )
```

```python
y_pred = model.predict(X_test)
```

```python
print( f'Оценка точности модели по MAPE: {round(mape(m.e**y_test, m.e**y_pred),4)}%')
```
