# Задание 1
Мы выпустили приложение Ленточка. После неплохого старта у нас начало сокращаться количество заказов из приложения.
Задача – выявить причины падения и предложить меры для решения этой ситуации. 
Необходимые данные можно придумать.

## Придумываю данные
В итоге, для анализа будет создан датасет из 10000 транзакций, содержающий слудющие столбцы:


**grocery** - сумма заказов из категори продукты за транзакцию

**household** - сумма заказов из категории хозтоваров

**electronics** - сумма заказов из категории электроники

**animals** - сумма заказов из категории "для животных"

**promo** - сумма промокода, используемого при оплате

**delivery_time** - время доставки

**date** - дата транзакции, в методических целях принимает значения от 1 до 30, имитируя месяц.

**location** - место доставки заказа, в методических целях принимает только значения Moscow и other

**device** - устройство, с которого был сделан заказ, в методических целях принимает значения только Iphone и Android


При генерации данных будет создана группа, отличная от остальных, которая и будет "причиной" падения заказов.

```python
import random as r
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def dgen_t(n):
    start=[]
    mid=[]
    fin=[]
    start_s=[]
    mid_s=[]
    fin_s=[]
    for i in range(int(n/3)):
        start.append(r.uniform(1,5))
        mid.append(r.uniform(4,5))
        fin.append(r.uniform(2,5))
    start_s = sorted(start)
    mid_s = sorted(mid)
    fin_s = sorted(fin, reverse=True)
    data = start_s+mid_s+fin_s
    data_ = [round(i,6) for i in data]
    return data_
```

```python
def plot_(data):
    df = pd.Series(data)
    sns.barplot(x = df.index, y = df.values)
```

```python
def dgen_o(n):
    start=[]
    mid=[]
    fin=[]
    start_s=[]
    mid_s=[]
    fin_s=[]
    for i in range(int(n/3)):
        start.append(r.uniform(1,5))
        mid.append(r.uniform(5,6))
        fin.append(r.uniform(6,7))
    start_s = sorted(start)
    mid_s = sorted(mid)
    fin_s = sorted(fin, reverse=True)
    data = start_s+mid_s+fin_s
    data_ = [round(i,6) for i in data]
    return data_
```

```python
def bin_f(n):
    f=[]
    for i in range(n):
        coeff = r.random()
        if coeff>0.3:
            f.append(1)
        else:
            f.append(0)
    return f
```

```python
categories = {'grocery':10**3, 'household':10**3, 'electronics':10**3, 'animals':10**3,'promo':10**2, 'delivery_time':10}
```

```python
feat_dict = {}
n = 10000
for item in categories:
    feat_dict[item] = [categories[item]*i for i in dgen_o(n)]
    if item in ['household','electronics']:
        feat_dict[item] = [categories[item]*i for i in dgen_t(n)]
df = pd.DataFrame.from_dict(feat_dict)    
```

```python
def df_plot(df,group,what,agg,xlabel,ylabel,title):
    data = df.groupby(group)[what].agg(agg)
    f, ax = plt.subplots(figsize=(20, 6))
    ax = sns.barplot(data.keys(),data,color = '#EEEBAA')
    ax.set_xticklabels(ax.get_xticks(),{'fontsize': 15,
     'fontweight': 50,
     'verticalalignment': 'top',
     'horizontalalignment': 'right'})
    ax.set_yticklabels(ax.get_yticks(),{'fontsize': 15,
     'fontweight': 50,
     'verticalalignment': 'top',
     'horizontalalignment': 'right'})
    ax.tick_params(axis='x', rotation=70)
    ax.grid(axis = 'y')
    ax.set_xlabel(xlabel,size = 20)
    ax.set_ylabel(ylabel,size = 20)
    ax.set_title(title, size = 25)
    return ax
```

```python
def date_f(n):
    f=[]
    chunk = int(n/30)
    day = 1
    c= int(n/chunk)
    for j in range(c):
        for i in range(chunk):
            f.append(day)
        day+=1
    diff = n-len(f)
    for i in range(diff):
        f.append(30)

    return f
```

```python
loc = {1:'Moscow', 0:'Other'}
dev = {1:'Iphone',0:'Android'}
```

```python
df.loc[:,'date'] = pd.Series(date_f(len(df)), index=df.index)
df.loc[:,'location'] = pd.Series(bin_f(len(df)), index=df.index)
df.loc[:,'device'] = pd.Series(bin_f(len(df)), index=df.index)
df['device'] = df['device'].apply(lambda x: dev[x])
df['location'] = df['location'].apply(lambda x: loc[x])
df['revenue'] = df[['grocery', 'household', 'electronics', 'animals']].sum(axis=1)
```

# Аналитика

Для начала обозначим проблему, со временем общая сумма заказов падает. Что можно видеть на графике.В начале имеется некий рост, который сменяется падением.

```python
df_plot(df,"date","revenue",'sum','День','Сумма заказов','Сумма заказов по дням')
```

Попробуем оценить влияние географической обстановки на сумму заказов.

```python
data=[]
days = [1,10,20,30]
for i in days:
    data.append([df[df['date']== i].groupby(['location'])['revenue'].sum()])
fig, ax = plt.subplots(2,2,figsize=[15,12])
for i in range(len(data)): 
    labels = data[i][0].keys()
    ax.flat[i].pie(x=data[i][0], autopct="%.1f%%", labels=labels, pctdistance=0.5,textprops = {"fontsize":15})
    ax.flat[i].set_title(f"Общая сумму заказов в день {days[i]}", fontsize=15);
```

Судя по графикам зависимость падения прибыли от региона не прослеживается, со временем выручка равномерно падает во вех регионах.

Аналогично посмотрим по девайсам, возможно за это можно будет зацепиться в дальнейшем.

```python
data=[]
for i in days:
    data.append([df[df['date']== i].groupby(['device'])['revenue'].sum()])
fig, ax = plt.subplots(2,2,figsize=[15,12])
for i in range(len(data)): 
    labels = data[i][0].keys()
    ax.flat[i].pie(x=data[i][0], autopct="%.1f%%", labels=labels, pctdistance=0.5,textprops = {"fontsize":15})
    ax.flat[i].set_title(f"Общая сумму заказов в день {days[i]}", fontsize=15);
```

Судя по всему, ситуация аналогичная, не прослеживается зависимость от девайсов со временем.

Рассмотрим остальные категории товаров.

```python
df_plot(df,"date","grocery",'sum','День','Сумма заказов, руб.','Сумма заказов продуктов по дням')
df_plot(df,"date","revenue",'sum','День','Сумма заказов, руб.','Сумма заказов по дням')
```

Продукты показывают не такое сильное падение, как общая выручка в целом. Возможно есть вклад, но он не велик.


```python
df_plot(df,"date","household",'sum','День','Сумма заказов, руб.','Сумма заказов хозтовары по дням')
df_plot(df,"date","revenue",'sum','День','Сумма заказов, руб.','Сумма заказов по дням')
```

Вот и первый кандидат на причину снижения количества заказов. Стоит обратить внимание на цепи поставок, промоакции и поставщиков хозтоваров.


```python
df_plot(df,"date","electronics",'sum','День','Сумма заказов, руб.','Сумма заказов электроники по дням')
df_plot(df,"date","revenue",'sum','День','Сумма заказов, руб.','Сумма заказов по дням')
```

Вот и второй кандидат, вносящик вклад в падение количества заказов. Аналогично, требуется дальнейший анализ этой области.

```python
df_plot(df,"date","animals",'sum','День','Сумма заказов, руб.','Сумма заказов товаров для животных по дням')
df_plot(df,"date","revenue",'sum','День','Сумма заказов, руб.','Сумма заказов по дням')
```

Товары для животных не вносят вклада в общие потери.

```python
df_plot(df,"date","promo",'mean','День','Среднее промо, руб.','Средняя величина промо услуг по дням')
```

Видимо, промо услуги идут вслед за заказами, чем меньше заказов, тем меньше промо услуг использует клиент. Возможно стоит увеличить вовлечение клиента за счет разнообразия промо акций.

```python
df_plot(df,"date","delivery_time",'mean','День','Время доставки, мин.','Среднее время доставки')
```

Среднее время доставки тоже снижается. Это может иметь эффект на желание клиента сделать повторный заказ.
